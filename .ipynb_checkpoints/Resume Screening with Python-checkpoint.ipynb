### Cleaning Data:
    1 URLs,
    2 hashtags,
    3 mentions,
    4 special letters,
    5 punctuations:

```python
import re

def clean_resume(text, remove_numbers=False, remove_domains=True):
    # Step 0: Validate input
    if not isinstance(text, str):
        return ""

    # Step 1: Lowercase
    text = text.lower()

    # Step 2: Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)

    # Step 3: Remove mentions and hashtags
    text = re.sub(r'@\w+|#\w+', ' ', text)

    # Step 4: Remove punctuation
    text = re.sub(r'[^\w\s]', ' ', text)

    # Step 5: Remove non-ASCII characters
    text = re.sub(r'[^\x00-\x7f]', ' ', text)

    # Step 6: Remove numbers (optional)
    if remove_numbers:
        text = re.sub(r'\d+', ' ', text)

    # Step 7: Remove domain-related words (like com, org, etc.)
    if remove_domains:
        domain_words = r'\b(com|org|net|edu|gov|in|co|io)\b'
        text = re.sub(domain_words, ' ', text)

    # Step 8: Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
## for example 
clean_resume("my #### $ #  #noorsaeed webiste like is this http://heloword and access it @gmain.com")
```

```python

df['Resume'] = df['Resume'].apply(lambda x: clean_resume(x))
```

```python
df['Resume'][0]
```

## Remove Stopwords

```python

import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

df['Resume'] = df['Resume'].apply(
    lambda x: ' '.join(
        lemmatizer.lemmatize(word)
        for word in x.split()
    )
)
```

```python
df['Resume'][0]
```

## Words into catagorical values 

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])
```

```python
df['Category'].unique()
```

```python
mapping = dict(zip(le.classes_, le.transform(le.classes_)))

print(mapping)
```

## Vactorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
requredTaxt  = tfidf.transform(df['Resume'])
```

```python
requredTaxt
```

## Splitting 

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(requredTaxt,df['Category'],test_size=0.2,random_state=42)
```

```python
X_train.shape,X_test.shape
```

## Now let’s train the model and print the classification report:

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Convert sparse matrices to dense only if required

X_train_dense = X_train.toarray() if hasattr(X_train, 'toarray') else X_train
X_test_dense = X_test.toarray() if hasattr(X_test, 'toarray') else X_test

models = {
"KNN": OneVsRestClassifier(KNeighborsClassifier()),
"SVC": OneVsRestClassifier(SVC()),
"Random Forest": RandomForestClassifier(random_state=42),
"Logistic Regression": LogisticRegression(max_iter=1000),
"Multinomial Naive Bayes": MultinomialNB(),
"Linear SVC": LinearSVC()
}

for name, model in models.items():

    # Models that require dense input
    if name in ["KNN", "SVC"]:
        model.fit(X_train_dense, y_train)
        y_pred = model.predict(X_test_dense)
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
    
    print(f"\n{'='*60}")
    print(f"{name}")
    print(f"{'='*60}")
    
    print("Accuracy Score:")
    print(f"{accuracy_score(y_test, y_pred):.4f}")
    
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

```

```python
models = {
    "KNN": OneVsRestClassifier(KNeighborsClassifier()),
"SVC": OneVsRestClassifier(SVC()),
"Random Forest": RandomForestClassifier(random_state=42),
"Logistic Regression": LogisticRegression(max_iter=1000),
"Multinomial Naive Bayes": MultinomialNB(),
"Linear SVC": LinearSVC()
}

for name, model in models.items():

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    print(f"\n{name}")
    print("-"*50)
    print("Accuracy:", accuracy_score(y_test, y_pred))
```

## Prediction System

```python
def pred(input_resume):
    
    # Clean resume
    cleaned_text = clean_resume(input_resume)
    
    # Vectorize
    vectorized_text = tfidf.transform([cleaned_text])
    
    # Predict
    predicted_category = model.predict(vectorized_text)
    
    # Decode label
    predicted_category_name = le.inverse_transform(predicted_category)
    
    print("Predicted Category:", predicted_category_name[0])
    
    return predicted_category_name[0]

```
